Jogadores confirmados.

Vendo as participações do Vini.JR na temporada de 2023-2024

```python
# Definindo o intervalo de datas
start_date = '2023-08-01'
end_date = '2024-07-31'

# Filtrar os eventos de Vinícius Jr. dentro desse intervalo de tempo
events_vinijr = games_events_df[(games_events_df["player_id"] == id_vinicius) & 
                                (games_events_df["date"] >= start_date) & 
                                (games_events_df["date"] <= end_date)]

# Exibir os primeiros 20 eventos filtrados
events_vinijr.head(20)
```

Vendo quantos gols o Vini fez no ano de 2024

```python
gols_vini = events_vinijr[events_vinijr['type'] == 'Goals']
gols_vini.head()
```

```python
num_gols_vini = len(gols_vini)
print(num_gols_vini)
```

Vendo as participações em gols do Rodri na temporada de 2023-2024

```python
# Definindo o intervalo de datas
start_date = '2023-08-01'
end_date = '2024-07-31'

# Filtrar os eventos de Rodri dentro desse intervalo de tempo
events_rodri = games_events_df[(games_events_df["player_id"] == id_rodri) & 
                                (games_events_df["date"] >= start_date) & 
                                (games_events_df["date"] <= end_date)]

# Exibir os primeiros 20 eventos filtrados
events_rodri.head(20)
```

Vendo número de gols do Rodri na temporada

```python
gols_rodri = events_rodri[events_rodri['type'] == 'Goals']
gols_rodri.head()
```

```python
num_gols_rodri = len(gols_rodri)
print(num_gols_rodri)
```

Criando um dataframe com todas os eventos ligados ao vini.jr e ao rodri na temporada de 2023-2024

```python
events_rodri_vinijr = pd.concat([events_rodri,events_vinijr]) #essa função concatena os dois dataframes
events_rodri_vinijr = events_rodri_vinijr.groupby(['player_id', 'type']).size().reset_index(name='count') # Contar o número de eventos de cada tipo para cada jogador


events_rodri_vinijr.head(20)
```

```python
# Criar um gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(data=events_rodri_vinijr, x='player_id', y='count', hue='type')

# Adicionar títulos e rótulos aos eixos
plt.title('Eventos por Jogador e Tipo')
plt.xlabel('ID do Jogador')
plt.ylabel('Número de Eventos')
plt.xticks(rotation=45)

plt.show()
```
