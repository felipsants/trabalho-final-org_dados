Vendo as participações do Vini.JR na temporada de 2023-2024

```python
# Definindo o intervalo de datas
start_date = '2023-08-01'
end_date = '2024-07-31'

# Filtrar os eventos de Vinícius Jr. dentro desse intervalo de tempo
events_vinijr = games_events_df[(games_events_df["player_id"] == id_vinicius) & 
                                (games_events_df["date"] >= start_date) & 
                                (games_events_df["date"] <= end_date)]

# Exibir os primeiros 20 eventos filtrados
events_vinijr.head(25)
```

Vendo quantos gols o Vini fez no ano de 2024

```python
appearances_vini = appearances_df[(appearances_df["player_id"] == id_vinicius) & 
                                (appearances_df["date"] >= start_date) & 
                                (appearances_df["date"] <= end_date)]
appearances_rodri = appearances_df[(appearances_df["player_id"] == id_rodri) & 
                                (appearances_df["date"] >= start_date) & 
                                (appearances_df["date"] <= end_date)]
```

```python
appearances_vini.info()
```

```python
appearances_vini.head()
```

```python
gols_vini = appearances_vini['goals'].sum()
```

```python
print(gols_vini)
```

Vendo as participações em gols do Rodri na temporada de 2023-2024

```python
# Definindo o intervalo de datas
start_date = '2023-08-01'
end_date = '2024-07-31'

# Filtrar os eventos de Rodri dentro desse intervalo de tempo
events_rodri = games_events_df[(games_events_df["player_id"] == id_rodri) & 
                                (games_events_df["date"] >= start_date) & 
                                (games_events_df["date"] <= end_date)]

# Exibir os primeiros 20 eventos filtrados
events_rodri.head(20)
```

Vendo número de gols do Rodri na temporada

```python
gols_rodri = appearances_rodri['goals'].sum()
```

```python
print(gols_rodri)
```

```python
assists_vini = appearances_vini['assists'].sum()
assists_rodri = appearances_rodri['assists'].sum()
participacao_gol_vini = gols_vini + assists_vini
participacao_gol_rodri = gols_rodri + assists_rodri
```

```python
print(assists_vini)
print(assists_rodri)
```

```python
linha_assists = pd.DataFrame({
    'player_id': ['Rodri','Vinicius Junior'],
    'type':['Assists','Assists'],
    'count':[assists_rodri,assists_vini]
})
```

Criando um dataframe com todas os eventos ligados ao vini.jr e ao rodri na temporada de 2023-2024

```python
events_rodri_vinijr = pd.concat([events_rodri,events_vinijr]) #essa função concatena os dois dataframes
events_rodri_vinijr = events_rodri_vinijr.groupby(['player_id', 'type']).size().reset_index(name='count') # Contar o número de eventos de cada tipo para cada jogador

events_rodri_vinijr = pd.concat([events_rodri_vinijr, linha_assists], ignore_index=True)

events_rodri_vinijr['player_id'] = events_rodri_vinijr['player_id'].replace({357565: 'Rodri', 371998: 'Vinicius Junior'})
events_rodri_vinijr['player_id'] = events_rodri_vinijr['player_id'].astype(str)
events_rodri_vinijr = events_rodri_vinijr.sort_values(by='player_id').reset_index(drop=True)
events_rodri_vinijr.head(20)
```

```python
# Criar um gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(data=events_rodri_vinijr, x='player_id', y='count', hue='type')

# Adicionar títulos e rótulos aos eixos
plt.title('Eventos por Jogador e Tipo')
plt.xlabel('ID do Jogador')
plt.ylabel('Número de Eventos')
plt.xticks(rotation=45)

plt.show()
```

Colentando os minutos jogados de cada jogador na temporada

```python
minutos_jogados_vini = appearances_vini['minutes_played'].sum()
minutos_jogados_rodri = appearances_rodri['minutes_played'].sum()
```

```python
print(minutos_jogados_vini)
print(minutos_jogados_rodri)
```

Calculando minutos para marcar um gol e minutos para participar de uma jogada de gol ( Gol ou Assistência para gol )

```python
minutos_para_marcar_vini = minutos_jogados_vini / gols_vini
minutos_para_marcar_rodri = minutos_jogados_rodri / gols_rodri
minutos_para_participar_gol_vini = minutos_jogados_vini / participacao_gol_vini
minutos_para_participar_gol_rodri = minutos_jogados_vini / participacao_gol_rodri
```

```python
print(minutos_para_marcar_rodri)
print(minutos_para_marcar_vini)
print(minutos_para_participar_gol_rodri)
print(minutos_para_participar_gol_vini)
```

```python
competitions_df.head()
```

```python
champions_linha = competitions_df.competition_code.apply(lambda name: name == 'uefa-champions-league')
champions_id = competitions_df[champions_linha]['competition_id'].values[0]
```

Coletando as métricas dos jogadores na competição Champions League 23/24

```python
appearances_vini_CL = appearances_df[(appearances_df["player_id"] == id_vinicius) & 
                                (appearances_df["date"] >= start_date) & 
                                (appearances_df["date"] <= end_date) &
                                (appearances_df["competition_id"] == champions_id)]
appearances_rodri_CL = appearances_df[(appearances_df["player_id"] == id_rodri) & 
                                (appearances_df["date"] >= start_date) & 
                                (appearances_df["date"] <= end_date) &
                                (appearances_df["competition_id"] == champions_id)]
```

```python
appearances_rodri_CL.head()
```

```python
appearances_vini_CL.head()
```

Minutos, gols e assistência na CL

```python
minutos_vini_CL = appearances_vini_CL['minutes_played'].sum()
minutos_rodri_CL = appearances_rodri_CL['minutes_played'].sum()
gols_vini_CL = appearances_vini_CL['goals'].sum()
gols_rodri_CL = appearances_rodri_CL['goals'].sum()
assists_vini_CL = appearances_vini_CL['assists'].sum()
assists_rodri_CL = appearances_rodri_CL['assists'].sum()
```

Participação em gols na CL

```python
participacao_gol_rodri_CL = gols_rodri_CL + assists_rodri_CL
participacao_gol_vini_CL = gols_vini_CL + assists_vini_CL
```

```python
print(minutos_vini_CL)
print(gols_vini_CL)
print(assists_vini_CL)
print(minutos_rodri_CL)
print(gols_rodri_CL)
print(assists_rodri_CL)
```

Minutos para marcar e participar de gol na CL

```python
minutos_para_marcar_vini_CL = minutos_vini_CL / gols_vini_CL
minutos_para_marcar_rodri_CL = minutos_rodri_CL / gols_rodri_CL
minutos_para_participar_gol_vini_CL = minutos_vini_CL / participacao_gol_vini_CL
minutos_para_participar_gol_rodri_CL = minutos_rodri_CL / participacao_gol_rodri_CL
```

```python
print(minutos_para_marcar_vini_CL)
print(minutos_para_participar_gol_vini_CL)
print(minutos_para_marcar_rodri_CL)
print(minutos_para_participar_gol_rodri_CL)
```

```python
df_eficiencia = pd.DataFrame({
    'Jogador': ['Vinicius Junior', 'Rodri'],
    'Minutos por Gol': [minutos_para_marcar_vini, minutos_para_marcar_rodri],
    'Minutos por Participação em Gol': [minutos_para_participar_gol_vini, minutos_para_participar_gol_rodri]
}).melt(id_vars='Jogador', var_name='Métrica', value_name='Minutos')

plt.figure(figsize=(10, 6))
sns.barplot(data=df_eficiencia, x='Jogador', y='Minutos', hue='Métrica')
plt.title('Eficiência de Vinicius Jr. e Rodri (Gols e Participações)')
plt.ylabel('Minutos')
plt.xlabel('Jogador')
plt.legend(title='Métrica')
plt.show()
```
